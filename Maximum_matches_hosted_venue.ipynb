```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IPL_Data/IPL_Matches_2008_2020.csv')
df
```

```python
# Find the Team who has played the maximum number of matches from 2008 - 2020 
```

```python
max_appearance =  df['venue'].value_counts().head(10)
```

```python
max_appearance
```

```python
# Convert the Series into a Dataframe
max_appearance_df = pd.DataFrame(max_appearance, columns=['venue'])
max_appearance_df
```

```python
# Convert the Series into a Dataframe
max_venue_played = df['venue'].value_counts().head(10)
max_appearance_df = pd.DataFrame(max_venue_played, columns=['venue'])
max_appearance_df['ground'] = max_appearance_df.index
max_appearance_df = max_appearance_df.reset_index()
max_appearance_df.drop(['index'], axis=1, inplace=True)
```

```python
max_appearance_df
```

```python
ax=sns.barplot(x='venue', y='ground', data=max_appearance_df)
ax.set(xlabel="Number of Matches", ylabel='Venue')
plt.grid() 
plt.show()
```
