```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IPL_Data/IPL_Matches_2008_2020.csv')
df
```

```python
df = df.replace('Pune Warriors', 'Rising Pune Supergiants', regex=True)
df = df.replace('Rising Pune Supergiants', 'Rising Pune Supergiant', regex=True)
df = df.replace('Rising Pune Supergiant', 'Rising Pune Supergiants', regex=True)
df = df.replace('Delhi Daredevils', 'Delhi Capitals', regex=True)
df = df.replace('Sunrisers Hyderabad', 'Deccan Chargers', regex=True)
```

```python
total_venues =  df['venue'].value_counts().head(10)
total_venues
```

## Check Win Percentage

```python
#Check Win Percentage

total_teams =  df['team1'].unique() 
total_venues =  df['venue'].value_counts()

venue_df = pd.DataFrame(columns=['team', 'venue', 'matches_played', 'won', 'win_percentage'])

for team in total_teams:
    for venue, count in total_venues.items():
        matches_played_df = df[ (df['venue'] == venue) & ( (df['team1'] == team) | (df['team2'] == team) )]
        total_matches_played = matches_played_df.shape[0]
        
        total_won =  matches_played_df[matches_played_df['winner'] == team].shape[0]
      
        win_percentage = 0 if total_matches_played == 0 else total_won*100/total_matches_played
      
        data = {'team':team, 'venue':venue,  'matches_played':matches_played_df.shape[0], 
                'won':total_won, 'win_percentage': win_percentage                
               }
        venue_df = venue_df.append(data, ignore_index=True)

# Pick Top 15 Venues where team have played a minimum of 10 matches        
venue_df[(venue_df['matches_played'] > 20) & (venue_df['win_percentage'] > 50)].sort_values('win_percentage', ascending=False).head(15)

```

## Check Loss Percentage

```python
total_teams =  df['team1'].unique() 
total_venues =  df['venue'].value_counts()

venue_df = pd.DataFrame(columns=['team', 'venue', 'matches_played', 'won', 'lost', 'win_percentage', 'loss_percentage'])

for team in total_teams:
    for venue, count in total_venues.items():
        matches_played_df = df[ (df['venue'] == venue) & ( (df['team1'] == team) | (df['team2'] == team) )]
        total_matches_played = matches_played_df.shape[0]
        
        total_won =  matches_played_df[matches_played_df['winner'] == team].shape[0]
        total_lost = total_matches_played -total_won
        
        win_percentage = 0 if total_matches_played == 0 else total_won*100/total_matches_played
        loss_percentage = 0 if total_matches_played == 0 else total_lost*100/total_matches_played
        
        data = {'team':team, 'venue':venue,  'matches_played':matches_played_df.shape[0], 
                'won':total_won, 'win_percentage': win_percentage,
                'lost':total_lost, 'loss_percentage': loss_percentage,
               }
        venue_df = venue_df.append(data, ignore_index=True)

# Pick Top 10 Venues where team have played a minimum of 10 matches        
venue_df[(venue_df['matches_played'] > 10) & (venue_df['loss_percentage'] > 50)].sort_values('loss_percentage', ascending=False).head(10)

```
